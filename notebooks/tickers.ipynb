```python
import pandas as pd
```

```python
fca_mobiliario=pd.read_csv('/home/eco/Metatrader_Cursor/join/fca_cia_aberta_valor_mobiliario_2025.csv',
                           sep = ';' , # Padrão é vírgula (,), trocamos para ponto e vírgula (; - csv!)
                      decimal = ',' , # Padrão do decimal é ponto (.), trocamos para vírgula (,)
                      encoding = 'ISO-8859-1') # Define a codificação dos caracteres para leitura latin1.)
#fca_mobiliario.head()
fca_mobiliario=(fca_mobiliario[["CNPJ_Companhia","Nome_Empresarial","Valor_Mobiliario","Codigo_Negociacao","Mercado"]]
                .rename(columns={"Nome_Empresarial":"DENOM_CIA","CNPJ_Companhia":"CNPJ_CIA"}).dropna()).drop_duplicates()
fca_mobiliario
```

```python
valores_unicos=fca_mobiliario["Mercado"].unique()
valores_unicos
```

```python
fca_mobiliario=fca_mobiliario[fca_mobiliario["Mercado"]=="Bolsa"]
fca_mobiliario
```

```python
fca_aberta=pd.read_csv('/home/eco/Metatrader_Cursor/join/fca_cia_aberta_2025.csv',
                           sep = ';' , # Padrão é vírgula (,), trocamos para ponto e vírgula (; - csv!)
                      decimal = ',' , # Padrão do decimal é ponto (.), trocamos para vírgula (,)
                      encoding = 'ISO-8859-1') # Define a codificação dos caracteres para leitura latin1.)
fca_aberta=fca_aberta[["CNPJ_CIA","DENOM_CIA","CD_CVM"]].drop_duplicates()
fca_aberta
```

```python
df_join=pd.merge(fca_aberta,fca_mobiliario,on=["CNPJ_CIA"],how="inner").drop_duplicates()
df_join=df_join.rename(columns={"DENOM_CIA_x":"DENOM_CIA"})

df_join=df_join.drop(columns=["Mercado","DENOM_CIA_y","CNPJ_CIA"],axis=1)
df_join
```

```python
bpa=pd.read_csv("/home/eco/Metatrader_Cursor/balance_sheet/dfp_cia_aberta_BPA_con_2020-2025.csv").drop_duplicates()
bpa=bpa[bpa["ORDEM_EXERC"]=="ÚLTIMO"]
bpa=bpa[["DT_REFER","DENOM_CIA","CD_CVM","CD_CONTA","DS_CONTA","VL_CONTA"]]
bpa
```

```python
df_final=pd.merge(df_join,bpa,on="CD_CVM",how="inner")
df_final=df_final.rename(columns={"DENOM_CIA_x":"DENOM_CIA"})
df_final=df_final.drop(columns="DENOM_CIA_y",axis=1).drop_duplicates()
df_final
```
